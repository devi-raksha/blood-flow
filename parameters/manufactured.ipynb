```
  set Density (rho)                           = 1
  set Elastic modulus of the vessel wall (mu) = 1
  set Reference cross-sectional area (a0)     = 3.14
  set Reference pressure (p0)                 = 1
  set Reference radius (r0)                   = 1
  set Tube law exponent (m)                   = 0.5
  set Viscosity coefficient (eta_c)           = 1
```

````python
from sympy import *
````

````python
rho, mu, a0, p0, r0, m, eta_c, x, t = symbols('rho mu a0 p0 r0 m eta_c x t')

A = a0*(1+(5.0e-3)*sin(2*3.14*(x-1.5*t)))
U = 1+0*x

A0 = A.subs({t:0})
U0 = U.subs({t:0})

P = p0 + mu*( (A/a0)**m - 1 )

flux_A = A*U
flux_U = U**2/2 + P/rho

RHS_A = simplify(diff(A, t) + diff(flux_A, x) )
RHS_U = simplify(diff(U, t) + diff(flux_U, x) + eta_c * U)
````

````python
expression = str(RHS_A) + '; ' + str(RHS_U)
expression =expression.replace('pi', 'PI').replace('**', '^')
````

````python
print( f"""
   set Exact solution = {A}; {U}
   set Initial condition = {A0}; {U0}
   set RHS expression = {expression}
"""
)
````
